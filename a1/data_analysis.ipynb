```python
import polars as pl
import scipy.stats as stats
```

```python
burger_orders = pl.read_csv('../data/burger_orders.csv')
burger_orders
```

```python
binom = stats.binom
poisson = stats.poisson

binom_cols = burger_orders['Bun', 'Patty', 'Bacon', 'Salat', 'Sauce']
poisson_cols = burger_orders['Gewuerz', 'Kaese', 'Gemuese']

print(f'Binom cols: {binom_cols.columns}')
for col_binom in binom_cols:
    bounds = [(0, 3), (0, 1)]
    res = stats.fit(binom, col_binom, bounds)
    print(f'{col_binom.name: <5}: n = {int(res.params[0])}, p = {res.params[1]:.2f}')

print(f'\nPoisson cols: {poisson_cols.columns}')
for col_pois in poisson_cols:
    bounds = [(0, 5), (0, 1)]
    res = stats.fit(poisson, col_pois, bounds)
    print(f'{col_pois.name: <7}: mu = {res.params[0]:.2f}')
```
